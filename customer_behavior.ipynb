```python
import pandas as pd
df=pd.read_csv("/customer_shopping_behavior.csv")
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

```python
#creating a new column age_group
def age_group(age):
  if age<30:
    return "Young Adult"
  elif age<40:
    return "Adult"
  elif age<60:
    return "Middle Aged"
  else:
    return "Senior"

df["age_group"]=df['age'].apply(age_group)
```

```python
df.head(10)
```

```python
# create column purchase_frequency_days

def freq(frequency_of_purchases):
  if frequency_of_purchases=="Fortnightly":
    return 14
  elif frequency_of_purchases=="Weekly":
    return 7
  elif frequency_of_purchases=="Monthly":
    return 30
  elif frequency_of_purchases=="Quarterly":
    return 15
  elif frequency_of_purchases=="Annually":
    return 365
  elif frequency_of_purchases=="Bi-Weekly":
    return 14
  else:
    return 90

df["purchase_frequency_days"]=df['frequency_of_purchases'].apply(freq)
```
